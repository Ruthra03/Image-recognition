```python
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Function to randomly select and display files from a folder
def display_random_images(folder_path, num_images):
    # List all files in the folder
    all_files = os.listdir(folder_path)
    # Shuffle the list of files randomly
    random.shuffle(all_files)
    # Select the first num_images files
    selected_files = all_files[:num_images]
    # Display the selected images
    for filename in selected_files:
        file_path = os.path.join(folder_path, filename)
        img = mpimg.imread(file_path)
        plt.imshow(img)
        plt.axis('off')  # Turn off axis
        plt.show()

# Define folder paths
folders = ['n03636648_new', 'n07734744_new', 'n11978233_new']
num_images_per_folder = 10

# Iterate through each folder
for folder in folders:
    folder_path = os.path.join('/content/drive/MyDrive/Dataset_image', folder)  # Replace '/path/to/parent/folder' with the actual parent folder path
    # Display random images from the folder
    display_random_images(folder_path, num_images_per_folder)
```
